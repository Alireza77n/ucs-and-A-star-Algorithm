```python
import re
from copy import deepcopy
import random
import time
```

```python
class ucs:
    def __init__(self,k):
        self.cart_num=k
        self.goal=[]
        self.goal_cost=[]
        self.row=[]
        self.priority_queue=[]
        self.visited={}
        self.G=0
        self.final_output=[]
        self.node_seen=0
    def insert(self,r):
        self.row.append(r.split())
        
    def goal_node(self):
        temp=[]
        sett=set()
        for j in range(len(self.row)):
            for k in range(len(self.row[j])):
                if self.row[j]!=['#']:
                        sett.add(re.findall(r'[A-Za-z]+|\d+', self.row[j][k])[1])  
                else:
                    pass
        for i in sett:
            for j in range(self.cart_num):
                temp.append(str(self.cart_num-j)+i)
            self.goal.append(temp) 
            temp=[]
        return self.goal
    
    def state(self,row=[]):
        flag=False
        temp_states=[]
        g=[]
        for i in range(len(row)):
            for j in range(len(row)):
                if i!=j:
                    if row[i]!=['#']:
                        if row[j]==['#']:
                            temp=deepcopy(row)
                            
                            cart=temp[i].pop()
                            if temp[i]==[]:
                                temp[i]=['#']
                            temp[j].pop()
                            temp[j].append(cart)
                            temp_states.append(temp)
                            temp=[]
                        if row[j]!=['#']:
                            if int(re.findall(r'[A-Za-z]+|\d+', row[i][-1])[0])<=int(re.findall(r'[A-Za-z]+|\d+', row[j][-1])[0]):
                                temp=deepcopy(row)
                                cart=temp[i].pop()
                                if temp[i]==[]:
                                    temp[i]=['#']
                                temp[j].append(cart)
                                temp_states.append(temp)
                                temp=[]
                                

                
        for i in range(len(temp_states)):
            g.append(1+self.G)
          
    #random.randint(self.G+1,self.G+10)
              

        for i in range(len(temp_states)):
            self.priority_queue.append((temp_states[i],g[i]))
            
       
        
        self.priority_queue.sort(key=lambda x:x[1])
        while str(self.priority_queue[0][0]) in list(self.visited.keys()):
            
            if self.priority_queue[0][1] < self.visited[str(self.priority_queue[0][0])]:
                
                self.visited.pop(str(self.priority_queue[0][0]))
                self.node_seen=self.node_seen-1
            else:
                self.priority_queue.pop(0)
            
        if  self.priority_queue==[]:
            print('Failur to find the answer')
            return

        a=self.priority_queue.pop(0)
        self.visited[str(a[0])]=a[1]
        self.node_seen=self.node_seen+1
        self.G=a[1]
        
        for i in range(len(self.goal)):
            if self.goal[i] in a[0]:
                flag=True
            else:
                flag=False
                break;
            
        if flag==True:
            
            self.goal_cost.append(self.G)
            self.goal_cost.sort()
            if self.priority_queue[0][1]>=self.goal_cost[0]:
                self.final_output=a[0]
                return
            
        self.state(a[0])     
    
```

```python
start = time.time()
q=ucs(2)
q.insert('1g')
q.insert('2y 2g')
q.insert('1y')
q.insert('#')
#q.insert('3y 3g 2y ')
#q.insert('2g 1g 1r 2r')
#q.insert('1y 3r')
#q.insert('#')
q.goal_node()
q.state(q.row)
print(q.final_output)
print('cost to reach the goal:',q.G)
print('number of visited node:',q.node_seen)
end = time.time()
print("The time of execution of above program is :", end-start)
```
