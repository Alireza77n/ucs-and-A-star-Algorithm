```python
import re
import time
from copy import deepcopy
import sys
```

```python
class Astar:
    def __init__(self,k):
        self.cart_num=k
        self.goal=[]
        self.row=[]
        self.priority_queue=[]
        self.visited={}
        self.G=0
        self.H=0
        self.final_output=[]
        self.node_seen=0
    def insert(self,r):
        self.row.append(r.split())
    def state(self,row=[]):
        temp_states=[]
        g=[]
        number=[]
        sett=set()
        h=[]
        f=[]
        for i in range(len(row)):
            for j in range(len(row)):
                if i!=j:
                    if row[i]!=['#']:
                        if row[j]==['#']:
                            temp=deepcopy(row)
                            
                            cart=temp[i].pop()
                            if temp[i]==[]:
                                temp[i]=['#']
                            temp[j].pop()
                            temp[j].append(cart)
                            temp_states.append(temp)
                            temp=[]
                        if row[j]!=['#']:
                            if int(re.findall(r'[A-Za-z]+|\d+', row[i][-1])[0])<=int(re.findall(r'[A-Za-z]+|\d+', row[j][-1])[0]):
                                temp=deepcopy(row)
                                cart=temp[i].pop()
                                if temp[i]==[]:
                                    temp[i]=['#']
                                temp[j].append(cart)
                                temp_states.append(temp)
                                temp=[]
                                
    
        for i in range(len(temp_states)):
            g.append(1+self.G)
            
            
        for i in range(len(temp_states)):
            temp=temp_states[i]
            for j in range(len(temp)):
                for k in range(len(temp[j])):
                    if temp[j]!=['#']:
                        number.append( int(re.findall(r'[A-Za-z]+|\d+', temp[j][k])[0]))
                        sett.add(re.findall(r'[A-Za-z]+|\d+', temp[j][k])[1])  
                if(len(sett)>1):
                    self.H=self.H+len(sett)
                else:
                    pass
                if len(number)!=0:
                    for i in range(len(number)):
                        if number[i]!=self.cart_num-i:
                            self.H=self.H+1
                        else:
                            pass
                   
                number=[]
                sett.clear()
                
            h.append(self.H)
            self.H=0    
        for i in range(len(h)):
            f.append(h[i]+g[i])

        for i in range(len(f)):
            self.priority_queue.append((temp_states[i],g[i],h[i],f[i]))
            
        self.priority_queue.sort(key=lambda x:x[3])
        
        while str(self.priority_queue[0][0]) in list(self.visited.keys()):
            if self.priority_queue[0][3] < self.visited[str(self.priority_queue[0][0])]:
                
                self.visited.pop(str(self.priority_queue[0][0]))
                self.node_seen=self.node_seen-1
            else:
                self.priority_queue.pop(0)
            
        if  self.priority_queue==[]:
            print('Failur to find the answer')
            return

        a=self.priority_queue.pop(0)
        self.node_seen=self.node_seen+1
        self.G=a[1]
        self.visited[str(a[0])]=a[3]
        if a[2]==0:
            self.goal.append(a[3])
            
        if self.goal!=[]:
            self.goal.sort()
            if a[2]==0:
                if self.priority_queue[0][3]>=self.goal[0]:
                    self.final_output=a[0]
                    return
            
        #print( a[0])        
        self.state(a[0])     
    
```

```python
start = time.time()
q=Astar(3)
#q.insert('1g')
#q.insert('2y 2g')
#q.insert('1y')
#q.insert('#')
q.insert('3y 3g 2y ')
q.insert('2g 1g 1r 2r')
q.insert('1y 3r')
q.insert('#')
q.state(q.row)
print(q.final_output)
print('cost to reach the goal:',q.G)
print('number of visited node:',q.node_seen)
end = time.time()
print("The time of execution of above program is :", end-start)
```
